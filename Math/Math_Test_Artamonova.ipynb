На вход программа должна принимать 2 строки.

Каждая строка содержит 3 действительных числа через пробел: коэффициенты a и b.

````python
#введенные строки необходимо преобразовать в числа с плавающей точкой
def str_list_to_float_list(line):
  coef_str_list = (line).split()
  coef_float_list = []
  for i in range(0, len(coef_str_list)):
    coef_float_list.append(float(coef_str_list[i]))
  return coef_float_list
````

````python
import numpy as np
from numpy import linalg as LA

def linal(first_line,second_line):
  coefficients_first = str_list_to_float_list(first_line)
  coefficients_second = str_list_to_float_list(second_line)
  a = [[coefficients_first[0],coefficients_first[1]],
       [coefficients_second[0],coefficients_second[1]]]
  b = [(coefficients_first[-1]),(coefficients_second[-1])]
  x = LA.solve(a,b)
  print(x[0],x[1])
 
linal(first_line,second_line)
````

````python
size = int(input())
````

````python
import numpy as np

def decorate_matrix(size):
  A = np.ones((size,size))
  A[1:-1,1:-1] = 0
  return A

print(decorate_matrix(size))
````

````python
def decorate_matrix_diagonal(size):
  B = np.eye(size)
  B[0:-1,0] = 1
  B[0:-1,-1] = 1
  B[0,0:-1] = 1
  B[-1,0:-1] = 1
  return B

print(decorate_matrix_diagonal(size))
````

# Задание 5

1000 человек бросают монетку. Каждый бросил 10 раз подряд.

 1. Какова вероятность того, что хотя бы у одного человека монетка выпала 10 раз подряд той же самой сторой (орлом или решкой)? 
 2. Какова вероятность того, что хотя бы у одного человека монетка выпала 8 раз подряд той же самой сторой (орлом или решкой)? 
 3. Какова вероятность того, что ровно у троих человек из 1000 монетка выпала 10 раз подряд той же самой сторой (орлом или решкой)?

````python
from math import factorial
````

````python
def bernulli(p, n, k):
  C_kn = factorial(n)/(factorial(k) * factorial(n - k))
  probability = C_kn * (p**k) * (1 - p)**(n - k)
  return probability

print(bernulli(0.5, 10, 10))
print(bernulli(0.5, 10, 8))
````

````python
probability_2 = 1-((1 - probability)**(1000))
probability_2
````

````python
def task5_1(all, throws, times):
  p = 0.5 #вероятность орла или решки
  probability_coin = bernulli(p,throws,times) #считаем вероятность выпадания монетки k раз из n одной и той же стороной
  probability = 1-((1 - round(probability_coin,4))**(all)) #обратная вероятность "хотя бы один из"
  return probability

print('Ответ к первой задаче:',task5_1(1000,10,10))
````

````python
def task5_2(all, throws, times):
  #для подсчета количества вожможных вариантов, объединяем 8 одинаковых выпадов 
  #монет в один объект. Считаем количество возможных позиций для объектов
  poses = throws - times + 1
  #считаем количество комбинаций
  coin_combination = factorial(poses)/(factorial(1) * factorial(poses - 1))
  #число всех комбинаций. возможные значения (орел, решка) в степени количества бросков
  all_combination = 2**10
  probability_coin = coin_combination/all_combination
  probability = 1-((1 - probability_coin)**(all))
  return probability

print('Ответ ко второй задаче:',task5_2(1000,10,8))
````

````python
import math

def task_5_3(all, human, throws, times):
  p = 0.5 #вероятность орла или решки
  probability_coin = round(bernulli(p,throws,times),3) #считаем вероятность выпадания монетки k раз из n одной и той же стороной
  #probability = bernulli(probability_coin, all, human) #считаем вероятность того,что у k людей из n монета выпадет одной и той же стороной
  #применяем пуассоновское приближение
  probability = (((all*probability_coin*2)**human)/factorial(human))*math.exp(-all*probability_coin*2)
  return probability

print('Ответ к третьей задаче:',task_5_3(1000,3,10,10))
````
