###  Описание данных
#### Таблица ```transactions.csv```
##### Описание
Таблица содержит историю транзакций клиентов банка за один год и три месяца.

##### Формат данных

```
customer_id,tr_datetime,mcc_code,tr_type,amount,term_id
111111,15 01:40:52,1111,1000,-5224,111111
111112,15 15:18:32,3333,2000,-100,11122233

```
##### Описание полей

```
customer_id,gender
111111,0
111112,1
```

## Задание 3
1. В датафрейме transactions найдите клиента с максимальной суммой приходов на карту
2. В датафрейме transactions найдите клиента с максимальной суммой расходов по карте
3. Найдите разницу для этих клиентов между суммой расходов и суммой приходов

````python
customer_id_amount = transactions[['customer_id','amount']]
````

````python
transactions_income = customer_id_amount[customer_id_amount.amount > 0]
transactions_outgo = customer_id_amount[customer_id_amount.amount < 0]
````

````python
transactions_income_total = transactions_income.groupby('customer_id').sum()
transactions_outgo_total = transactions_outgo.groupby('customer_id').sum()
````

````python
max_income = transactions_income_total['amount'].max()
max_outgo = transactions_outgo_total['amount'].min()
````

````python
max_income_customer = transactions_income_total[transactions_income_total.amount == max_income]
print('Задание 3.1:', max_income_customer.index[0])
````

````python
max_outgo_customer = transactions_outgo_total[transactions_outgo_total.amount == max_outgo]
print('Задание 3.2:', max_outgo_customer.index[0])
````

````python
difference = (abs(max_income) - abs(max_outgo))
print('Задание 3.3:', difference)
````

Вывод: сумма расходов превышает сумму приходов

## Задание 6

1. По всем типам транзакций рассчитайте максимальную сумму прихода на карту (из строго положительных сумм по столбцу amount) отдельно для мужчин и женщин (назовите ее "max_income"). Оставьте по 10 транзакций для мужчин и для женщин, наименьших среди всех транзакций по полученным значениям "max_income". 
2. Выделите среди них те, которые встречаются одновременно и у мужчин, и у женщин

````python
income_transactions = transactions[transactions.amount > 0]
groups_gender_tr = income_transactions.groupby(['gender','tr_type']).amount.sum().unstack()
````

````python
print('Задание 6.1:')
max_income_for_0 = groups_gender_tr.T.nsmallest(10, 0)[0]
max_income_for_0
````

````python
print('Задание 6.1:')
max_income_for_1 = groups_gender_tr.T.nsmallest(10, 1)[1]
max_income_for_1
````

````python
print('Задание 6.2:')
pd.merge(max_income_for_0, max_income_for_1, on='tr_type')
````

## Задание 8

1. Из поля tr_datetime выделите час tr_hour, в который произошла транзакция, как первые 2 цифры до ":". (\**)
2. Посчитайте ночные поступления для мужчин и женщин. Ночные - это примерно 00-06 часов. 

````python
transactions.insert(1, 'tr_hour', transactions['tr_datetime'].apply(lambda x: x.split()[1].split(':')[0]))
print('Задание 8.1:')
transactions['tr_hour'] = pd.to_numeric(transactions['tr_hour'])
transactions
````

````python
night_transaction_0 = transactions[(transactions['gender'] == 0) & 
                                   (transactions['tr_hour'] < 6) &
                                   (transactions['amount'] >= 0)]
night_transaction_1 = transactions[(transactions['gender'] == 1) & 
                                   (transactions['tr_hour'] < 6) &
                                   (transactions['amount'] >= 0)]
````

````python
print('Задание 8.2:')
print('Количество ночных транзакций для пола 0:', night_transaction_0.amount.sum())
print('Количество ночных транзакций для пола 1:', night_transaction_1.amount.sum())
````
