### Задача 3

1). Создайте тензор *целых чисел* `images` размерности (100, 200, 200, 3) (можно интерпретировать это как 100 картинок размера 200х200 с тремя цветовыми каналами, то есть 100 цветных изображений 200х200). Заполните его нулями  

2). Сделайте так, чтобы у $i$-ой по порядку картинки была нарисована белая полоса толщиной два пикселя в строках, которые соответствуют номеру картинки. Например, у 3-ей по порядку картинки белая полоска будет занимать 6 и 7 строки, у 99 - 198 и 199. Сделать белую строку можно, записав в ячейки тензора число 255 (по всем трём каналам). Выведите несколько примеров с помощью `pylot`  

3). Посчитайте среднее тензора `images` по 1-ой оси (по сути - средняя картинка), умножьте полученный тензор на 70. Нарисуйте с помощью `pyplot`, должна получиться серая картинка (при взятии среднего нужно привести тензор к float с помощью `.float()`, при отрисовке обратно к int с помощью `.int()`)

```python
# Ваш код здесь
# 1. Создайте тензор целых чисел images размерности (100, 200, 200, 3). Заполните его нулями
images = torch.IntTensor(100, 200, 200,3).zero_()
```

```python
# 2. Сделайте так, чтобы у  i -ой по порядку картинки была нарисована белая полоса толщиной два пикселя в строках, которые соответствуют номеру картинки.
def white_stripes(images):
  for i in range(images.shape[0]):
    images[i,(i*2),:,:] = 255
    images[i,(i*2+1),:,:] = 255
  return images
```

```python
new_image = white_stripes(images)
```

```python
# 2. Выведите несколько примеров с помощью pylot
import matplotlib.pyplot as plt
plt.imshow(new_image[40,:,:,:]);
```

```python
# 3. Посчитайте среднее тензора images по 1-ой оси, умножьте полученный тензор на 70. Нарисуйте с помощью pyplot
i_mean = new_image.type_as(torch.FloatTensor()).mean(0)
i_mean = i_mean.mul(70)
i_mean = i_mean.type_as(torch.IntTensor())
import matplotlib.pyplot as plt
plt.imshow(i_mean);
```

### Задача 4

- В Colab нажмите в меню сверху: `Runtime` -> `Change Runtime Type` -> `GPU`, где изначально стоит `None` (`Среда выполнение` - `Сменить среду выполнения`)
- Объявите тензор `a` размера (2, 3, 4) и тензор `b` размера (1, 8, 3) на GPU, иницилизируйте их случайно равномерно (`.uniform_()`)
- Затем измените форму тензора `b`, чтобы она совпадала с формой тензора `a`, получите тензор `c`  
- Переместите `c` на CPU, переместите `a` на CPU  
- Оберните их в `Variable()`
- Объявите тензор `L = torch.mean((c - a) `**` 2)` и посчитайте градиент `L` по `c` ( то есть $\frac{\partial{L}}{\partial{c}})$
- Выведите градиент `L` по `c`

```python
# Ваш код здесь
# 1. Объявите тензор a размера (2, 3, 4) и тензор b размера (1, 8, 3) на GPU, иницилизируйте их случайно равномерно (.uniform_())
a = torch.FloatTensor(2,3,4).uniform_()
b = torch.FloatTensor(1,8,3).uniform_()
```

```python
# 2. Затем измените форму тензора b, чтобы она совпадала с формой тензора a, получите тензор c
c = b.view(2,3,4)
```

```python
# 3. Переместите c на CPU, переместите a на CPU
c = c.cpu()
a = a.cpu()
```

```python
# 4. Оберните их в Variable()
from torch.autograd import Variable
a = Variable(a, requires_grad=True)
c = Variable(c, requires_grad=True)
```

```python
# 5. Объявите тензор L = torch.mean((c - a)**2) и посчитайте градиент L по c
L = torch.mean((c - a)**2)
L.backward()
```

```python
# 6. Выведите градиент L по c
c.grad
```
